# Does it matter what log transformation you use?


````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Set up variables, x is 1-99 and y is e^x
x = list(np.linspace(0,10,100))
y= [np.exp(i) for i in x]
````

# Plot _x_  and  *e<sup>x</sup>*

````python

fig, ax = plt.subplots(figsize=(8,6))
# Set background of area behind axis labels
fig.patch.set_facecolor('white')
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('$e^x$',fontsize=20,rotation=0,)

plt.tight_layout()
ax.grid(1)
plt.plot(x,y)
# plt.plot(actuals)
plt.title('$e^x$ vs. $x$',fontsize = 30, pad=10)
plt.axhline(y=0, color='black');
ax.yaxis.set_label_coords(-0.1,.5)

plt.savefig('graphs/e_vs_x.png', dpi=300, bbox_inches = "tight")
# ax.title.set_label_coords(-0.1,.5)
````

# Plot _x_ and log<sub>10</sub> ( *e<sup>x</sup>* ) and ln(*e<sup>x</sup>*)

````python


fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,6))
# Set background of area behind axis labels
fig.patch.set_facecolor('white')
ax1.set_xlabel('x', fontsize=20)
ax1.set_ylabel('$log(e^x)$',fontsize=20,)
ax1.axhline(y=0, color='black');
plt.tight_layout()
ax1.grid(1)
ax1.plot(x,np.log10(y))
ax1.set_title('$log(e^x)$ vs. $x$', fontsize = 30, pad=10)


ax2.set_xlabel('x', fontsize=20)
ax2.set_ylabel('$ln(e^x)$',fontsize=20,)
plt.tight_layout()
ax2.grid(1)
ax2.plot(x,np.log(y))
ax2.set_title('$ln(e^x)$ vs. $x$', fontsize = 30, pad=10)
ax2.axhline(y=0, color='black');

plt.savefig('graphs/logs_comparisons.png', dpi=300, bbox_inches = "tight")
````

They are both linear! The only thing that changes between the two log bases is the slope, which would have no effect on data modeling. See the derivation in the post for why this is the case!

log_a(b) = \frac{log_x(a)}{log_x(b)}

Substitute a=e and b=y
log_e(y) = \frac{log_{10}(y)}{log_{10}(e)}

ln(y) = \frac{log(y)}{log(e)}
